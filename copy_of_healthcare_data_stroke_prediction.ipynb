```python
Q1 = sds['bmi'].quantile(0.25)
Q3 = sds['bmi'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
print(Q3)
print(Q1)
```

now we try removing all the outliers from the dataframe

```python
sds_numeric_remove_outlier = sds['bmi'][~((sds['bmi'] < (Q1 - 1.5 * IQR)) |(sds['bmi'] > (Q3 + 1.5 * IQR)))]
sds_numeric_remove_outlier.shape
```

```python
sds_numeric_remove_outlier[:5]
```

```python
sds_new=pd.DataFrame(data=sds_numeric_remove_outlier)
```
